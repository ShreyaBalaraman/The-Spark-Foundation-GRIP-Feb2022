Data Visualization 

```python
plt.figure(figsize=(15,10))
country=data_ter1['country_txt'].value_counts().index
sns.set_theme(style="whitegrid")
sns.countplot(data_ter1['country_txt'],order=country[:10])
plt.xlabel("Countries")
plt.ylabel("Count of terrorist attacks")
plt.title("Top 10 countries having highest number of terrorist attacks")
```

```python
plt.figure(figsize=(25,10))
region=data_ter1['region_txt'].value_counts().index
sns.set_theme(style="whitegrid")
sns.countplot(data_ter1['region_txt'],order=region[:10])
plt.xlabel("Regions")
plt.ylabel("Count of terrorist attacks")
plt.title("Top 10 regions having highest number of terrorist attacks")
```

```python
plt.figure(figsize=(15,10))
city=data_ter1['city'].value_counts().index
sns.set_theme(style="whitegrid")
sns.countplot(data_ter1['city'],order=city[:10])
plt.xlabel("cities")
plt.ylabel("Count of terrorist attacks")
plt.title("Top cities having highest number of terrorist attacks")
```

```python
plt.figure(figsize=(15,10))
year=data_ter1['iyear'].value_counts().index
sns.set_theme(style="whitegrid")
sns.countplot(data_ter1['iyear'],order=year[:10])
plt.xlabel("Year")
plt.ylabel("Count of terrorist attacks")
plt.title("The year in which the top 10 countries having highest number of terrorist attacks")
```

```python
plt.figure(figsize=(25,10))
target=data_ter1['targtype1_txt'].value_counts().index
sns.set_theme(style="whitegrid")
sns.countplot(data_ter1['targtype1_txt'],order=target[:10])
plt.xlabel("Type of property targeted")
plt.ylabel("Count of terrorist attacks")
plt.title("Top 10 categories targeted by terrorists")
```

```python
plt.figure(figsize=(50,15))
wepon=data_ter1['weaptype1_txt'].value_counts().index
sns.set_theme(style="whitegrid")
sns.countplot(data_ter1['weaptype1_txt'],order=wepon[:10])
plt.xlabel("Weapons")
plt.ylabel("Number of times the weapons been used by terrorists.")
plt.title("Top 5 categories of weapons used by the terrorists.")
```

```python
plt.figure(figsize=(20,10))
group_name=data_ter1['gname'].value_counts().index
sns.set_theme(style="whitegrid")
sns.countplot(data_ter1['gname'],order=group_name[:5])
plt.xlabel("Gangsters")
plt.ylabel("Count of gangsters attacks")
plt.title("Top 5 categories of terrorist groups")
```
